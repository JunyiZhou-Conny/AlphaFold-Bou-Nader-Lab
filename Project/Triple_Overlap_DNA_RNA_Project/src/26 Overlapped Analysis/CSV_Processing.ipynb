```python
import csv
import re

csv_path = "/Users/conny/Desktop/AlphaFold/Summer Project/src/26 Overlapped Analysis/26_overlaps_summary.csv"

# Regex to extract accession after the last _number_
pattern = re.compile(r'_\d+_([0-9a-zA-Z]+)_')

# Read and update rows in memory
with open(csv_path, newline='') as infile:
    reader = csv.reader(infile)
    rows = list(reader)

header = rows[0]
data_rows = rows[1:]

for row in data_rows:
    match = pattern.search(row[0])
    if match:
        row[0] = match.group(1)

# Write back to the same file
with open(csv_path, 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header)
    writer.writerows(data_rows)

print(f"Updated {csv_path} in place!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
csv_path = "/Users/conny/Desktop/AlphaFold/Summer Project/src/26 Overlapped Analysis/26_overlaps_summary.csv"
df = pd.read_csv(csv_path)

# Make a scatter plot
plt.figure(figsize=(8,6))
sc = plt.scatter(df['iptm'], df['ptm'], c=df['ranking_score'], cmap='viridis', s=80, edgecolor='k')
plt.colorbar(sc, label='Ranking Score')
plt.xlabel('iptm')
plt.ylabel('ptm')
plt.title('iptm vs ptm colored by ranking_score')
plt.grid(True)
plt.tight_layout()
plt.show()
```
